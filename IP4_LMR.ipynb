#### Question 5

```python
# Remove all the symbols (\$\,)from the 'Price' column and update it to  64-bit int numbers using astype(‘int64’) 
```

```python
car_sales.info()
```

```python
car_sales['Price']=car_sales['Price'].replace("[$,]", "", regex=True).astype(int)
car_sales.info()
```

```python
car_sales["Price"]=car_sales["Price"].apply(np.int64)
car_sales.info()
```

#### Question 6

```python
# add a column called 'Total Sales' which cumulatively adds the 'Price' column look up the cumsum() function
car_sales['Total Sales']=car_sales['Price'].cumsum()
car_sales.head()
```

#### Question 7

```python
# Add a column called 'Sale Date' which lists a series of successive dates starting from today
car_sales['Sale Date'] = pd.date_range('today', periods=len(car_sales)).normalize()
```

```python
# Find out Pearson correlation between 'Total Sales','Price','Odometer (KM)'. Comment on the correlation 
heat_map_corr=car_sales[['Total Sales','Price','Odometer (KM)']].corr(method='pearson')
print(heat_map_corr)
cols=['Total Sales',
      'Price','Odometer (KM)']
heat_map=sns.heatmap(heat_map_corr,
                     annot=True,
                     xticklabels=cols,
                     yticklabels=cols)
```
